```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
```

```python
crf=pd.read_csv("Company_Data.csv")
crf
```

```python
label_encoder=preprocessing.LabelEncoder()
crf["ShelveLoc"]=label_encoder.fit_transform(crf["ShelveLoc"])
crf["Urban"]=label_encoder.fit_transform(crf["Urban"])
crf["US"]=label_encoder.fit_transform(crf["US"])
crf
```

```python
array=crf.values
array
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(crf.iloc[:,:])
df_norm
```

```python
array=crf.values
x=df_norm.iloc[:,1:11]
y=df_norm["Sales"]
```

```python
num_trees = 500
max_features = 4
kfold = KFold(n_splits=10, random_state=7,shuffle=True)
```

```python
model = RandomForestRegressor(n_estimators=num_trees, max_features=max_features)
```

```python
results = cross_val_score(model, x, y, cv=kfold,scoring='neg_mean_squared_error')
print(results.mean()) 
```
